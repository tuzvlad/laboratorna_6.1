```python
name = input("Введіть своє ім'я: ")
print("Привіт, " + name + "! ")

surname = input("Введіть своє прізвище: ")
print ("Твоє прізвище - " + surname +" ")

age = input("Введіть свій вік: ")
print( "Тобі - "+ age +" ")

print("Ласкаво просимо до світу Python!")
```

Користувач вводить своє:

*   ім'я
*   прізвище
*   вік

у відповідь на відповідні запитання, і програма виводить ці дані разом із вітанням.

```python
# Введення списку речень
речення = input("Введіть список речень, розділених комами: ").split(',')

# Знаходження найдовшого речення
найдовше_речення = max(речення, key=lambda x: len(x.strip()))

# Виведення найдовшого речення
print("Найдовше речення: ", найдовше_речення.strip())
```

**Речення** = input("Введіть список речень, розділених комами: ").split(',')

---



**Програма** запитує користувача ввести список речень, розділених комами.
input() використовується для отримання введення від користувача. Рядок, який вводить користувач, розділяється за допомогою ком в функції split(','), створюючи таким чином список речень.
найдовше_речення = max(речення, key=lambda x: len(x.strip()))


---


У *цьому* рядку використовується функція max(), щоб знайти найдовше речення в списку речення. Параметр key вказує, яка властивість об'єктів списку (у цьому випадку, довжина речення після видалення зайвих пробілів з початку і кінця рядка за допомогою strip()) повинна використовуватися для порівняння речень і знаходження найбільшого.
print("Найдовше речення: ", найдовше_речення.strip())


---


*Найзрештою*, програма виводить найдовше речення, видаляючи зайві пробіли з початку і кінця рядка за допомогою strip().
 Результат виводиться у вигляді рядка, який містить слово "Найдовше речення:" і саме найдовше речення з введеного користувачем списку.

```python
# Введення рядка
рядок = input("Введіть рядок: ")

# Конвертація рядка в великі літери та виведення результату
великі_літери = рядок.upper()
print("Тільки великі літери: ", великі_літери)
```





✈  Рядок = *input*("Введіть рядок: ")
Цей рядок запитує користувача ввести рядок з клавіатури.
Функція *input*()
виводить повідомлення "Введіть рядок: " і чекає, доки користувач введе рядок. Потім введений рядок зберігається у змінній рядок.
великі_літери = рядок.upper()

✈ У цьому рядку використовується метод *upper*(), який викликається на рядку рядок.
Цей метод перетворює всі літери рядка великими літерами (*uppercase*) і зберігає результат у змінній великі_літери.
*print*("Тільки великі літери: ", великі_літери)

✈ Нарешті, ця інструкція виводить результат - рядок, який містить текст "Тільки великі літери: " разом із великими літерами, які були введені користувачем. Змінна великі_літери містить перетворений рядок, і її значення виводиться на екран.
